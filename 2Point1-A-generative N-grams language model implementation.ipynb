```python
import os
import nltk

# 1) Set the REAL path to the folder containing hamlet.txt
corpus_root = r"C:\Users\HADI\Downloads"   # change if needed
file_id = "hamlet.txt"

# 2) Validate file exists
full_path = os.path.join(corpus_root, file_id)
if not os.path.exists(full_path):
    raise FileNotFoundError(
        f"File not found!\nExpected: {full_path}\n\n"
        f"Fix: Put hamlet.txt there or update corpus_root."
    )

# 3) Download NLTK tokenizer resources
download_dir = r"C:\Users\HADI\nltk_data"
nltk.download("punkt_tab", download_dir=download_dir)
nltk.download("punkt", download_dir=download_dir)

if download_dir not in nltk.data.path:
    nltk.data.path.append(download_dir)

from nltk.corpus.reader import PlaintextCorpusReader
from nltk.util import everygrams
from nltk.lm.preprocessing import pad_both_ends, flatten, padded_everygram_pipeline
from nltk.lm import MLE



# 4) Create corpus from .txt files in the chosen folder
my_corpus = PlaintextCorpusReader(corpus_root, r".*\.txt")

# 5) Read and materialize sentences
sentences = list(my_corpus.sents(fileids=file_id))

# Count sentences
print("Number of sentences:", len(sentences))


# Print sentences
for sent in sentences:
    print(sent)

# Pad a specific sentence with start/end markers
padded_bigrams = list(pad_both_ends(sentences[1104], n=2))
print(list(everygrams(padded_bigrams, max_len=3)))

# Flatten padded sentences
flattened_tokens = list(
    flatten(
        pad_both_ends(sent, n=2)
        for sent in sentences
    )
)
print(flattened_tokens)

# Create training data and vocab for a trigram model
train, vocab = padded_everygram_pipeline(3, sentences)

# Train Maximum Likelihood Estimation model
lm = MLE(3)

print(len(lm.vocab))  # likely 0 before fitting

lm.fit(train, vocab)

print(lm.vocab)
print(len(lm.vocab))

# Generate text conditioned on previous words
print(lm.generate(6, ["to", "be"]))

# Vocabulary lookup examples
print(list(lm.vocab.lookup(sentences[1104])))
print(list(lm.vocab.lookup(["aliens", "from", "Mars"])))

# N-gram counts
print(lm.counts)
print(lm.counts[["to"]]["be"])

# Probabilities
print(lm.score("be"))
print(lm.score("be", ["to"]))
print(lm.score("be", ["not", "to"]))

# Log probabilities
print(lm.logscore("be"))
print(lm.logscore("be", ["to"]))
print(lm.logscore("be", ["not", "to"]))

# Entropy and perplexity
test = [("to", "be"), ("or", "not"), ("to", "be")]
print("lm.entropy(test):",lm.entropy(test))
print("lm.perplexity(test):",lm.perplexity(test))
```
